```python
import pandas as pd

dataset=pd.read_csv('/content/drive/MyDrive/machine learning/DAY8 (Unsupervised Learning)/iris (1).csv').values
print(dataset)
```

```python
data=dataset[:,0:4]
print(data)
```

```python
from sklearn.cluster import MeanShift

model=MeanShift(bandwidth=0.85) #K value
model.fit(data)
labels=model.labels_

print('no. of cluster:',len(model.cluster_centers_))
print('cluster centers:',model.cluster_centers_)
```

```python
labels = model.labels_

print("0-50:",labels[:50]) # mul data 50, labels pennanaw.
print("===================================================================")
print("50-100:",labels[50:100])
print("===================================================================")
print("100-150:",labels[100:150])
```
